```python
N = 1000
num_iter = 100

x = np.linspace(-2,2,N)
y = np.linspace(-2,2,N)
[X,Y] = np.meshgrid(x,y)
#data = np.ones((N,N))
data = np.full((N,N), num_iter)

c = complex(x[0],y[0])
z = complex(0)

for i in range(N):
    for j in range(N):
        c = complex(x[i],y[j])
        z = complex(0)
        for k in range(num_iter):
            z_new = z**2 + c
            if abs(z_new) > 2:
                #data[j,i] = 0
                data[j,i] = k
                break
            z = z_new
        
        
# pcolormesh(X,Y,data)
# title("Mandelbrot Set")
# xlabel("Real Part")
# ylabel("Imaginary Part")
# savefig("Mandelbrotset.jpg")
# show()
```
